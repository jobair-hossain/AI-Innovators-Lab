```python

# ============================================================  # This long line is only a visual separator to make the notebook easier to read.
# AI Innovators Lab: Handwritten Digit Recognition              # This is the title of today's AI project.
# Goal: Train an AI model that can recognize handwritten digits  # Our AI will look at a handwritten number image.
#       from 0 to 9 using the MNIST dataset.                    # The answer will be one digit: 0, 1, 2, 3, 4, 5, 6, 7, 8, or 9.
# Dataset: MNIST                                                # MNIST is a famous collection of handwritten digit images.
# ============================================================  # This separator shows that the introduction section is finished.

print("Welcome to the Handwritten Digit Recognition Project!")  # Show a welcome message on the screen.
print("Goal: Build an AI model that can recognize handwritten numbers from 0 to 9.")  # Tell students what the program will do.

# ============================================================  # Start a new section of the notebook.
# Step 1: Import required libraries                             # Libraries give Python extra tools that are not built in by default.
# ============================================================  # End of the Step 1 heading.

import numpy as np                         # NumPy helps us work with lists of numbers, especially image pixel values.
import matplotlib.pyplot as plt            # Matplotlib helps us draw pictures, charts, and graphs.
import tensorflow as tf                    # TensorFlow is the AI library we use to build and train the neural network.
from tensorflow.keras import layers, models # Keras gives us simple building blocks such as layers and models.

print("Libraries loaded successfully!")   # Print this message so we know the imports worked without errors.
```

```python

# ============================================================  # Visual separator for the next section.
# Step 2: Load the MNIST dataset                                # In this step, we bring the handwriting data into the notebook.
# MNIST contains thousands of handwritten digit images.          # The dataset has many examples so the AI can learn patterns.
# Each image is 28 x 28 pixels.                                 # Every digit image is very small: 28 pixels wide and 28 pixels tall.
# ============================================================  # End of the Step 2 heading.

mnist = tf.keras.datasets.mnist             # Store the MNIST dataset loader in a variable named mnist.

(x_train, y_train), (x_test, y_test) = mnist.load_data()  # Download/load the data and split it into training and testing parts.

print("Dataset loaded successfully!")       # Tell us the dataset was loaded.
print("Training images:", x_train.shape)    # Show how many training images we have and the size of each image.
print("Testing images:", x_test.shape)      # Show how many testing images we have and the size of each image.
```

```python

# ============================================================  # Visual separator for the next section.
# Step 3: Display sample images                                 # Before training, we look at a few examples from the dataset.
# This helps students see what the dataset looks like.           # Seeing examples makes the AI problem easier to understand.
# ============================================================  # End of the Step 3 heading.

plt.figure(figsize=(10, 4))                 # Create a blank drawing area that is 10 inches wide and 4 inches tall.

for i in range(10):                         # Repeat the following indented lines 10 times, using i = 0 through 9.
    plt.subplot(2, 5, i + 1)                # Make a grid with 2 rows and 5 columns, then choose the next box in the grid.
    plt.imshow(x_train[i], cmap="gray")     # Show the i-th training image in black and white.
    plt.title(f"Label: {y_train[i]}")       # Put the correct answer, called the label, above the image.
    plt.axis("off")                         # Hide the x-axis and y-axis numbers because they are not needed here.

plt.tight_layout()                          # Adjust spacing so titles and images do not overlap.
plt.show()                                  # Display the finished group of images.
```

```python

# ============================================================  # Visual separator for the next section.
# Step 4: Prepare the data                                      # AI models usually need data to be cleaned or scaled before training.
# Neural networks work better when pixel values are scaled.      # Smaller values help the model learn more smoothly.
# Original pixel values are from 0 to 255.                       # 0 means black, 255 means white, and numbers between them are shades of gray.
# We convert them to values between 0 and 1.                     # Dividing by 255 changes the range from 0-255 to 0-1.
# ============================================================  # End of the Step 4 heading.

x_train = x_train / 255.0                   # Scale all training image pixels so every value is between 0 and 1.
x_test = x_test / 255.0                     # Scale all testing image pixels in the same way.

print("Data normalized successfully!")      # Print a message after the pixel values have been scaled.
```

```python

# ============================================================  # Visual separator for the next section.
# Step 5: Build the AI model                                    # Now we create the neural network that will learn from the images.
# This is a simple neural network.                              # A neural network is a program that learns patterns from examples.
# Flatten converts a 28 x 28 image into a list of 784 numbers.  # 28 times 28 equals 784, so each pixel becomes one number in a long list.
# Dense layers help the model learn patterns.                   # Dense layers connect many numbers together to discover useful patterns.
# ============================================================  # End of the Step 5 heading.

model = models.Sequential([                 # Create a model where layers are placed one after another in order.
    layers.Flatten(input_shape=(28, 28)),   # Turn each 28 x 28 image into one long list of 784 pixel values.
    layers.Dense(128, activation="relu"),   # Add a layer with 128 learning units; ReLU helps the model learn non-simple patterns.
    layers.Dense(64, activation="relu"),    # Add another layer with 64 learning units to learn more digit patterns.
    layers.Dense(10, activation="softmax")  # Add the final layer with 10 outputs, one for each digit from 0 to 9.
])                                          # End the list of model layers.

model.compile(                              # Prepare the model for training by choosing how it will learn.
    optimizer="adam",                       # Adam is the method that updates the model to reduce mistakes.
    loss="sparse_categorical_crossentropy", # This loss function measures how wrong the model is for number classification.
    metrics=["accuracy"]                    # Accuracy tells us the percentage of images the model predicts correctly.
)                                           # Finish compiling the model.

model.summary()                             # Print a table that shows the layers and number of trainable values in the model.
```

```python

# ============================================================  # Visual separator for the next section.
# Step 6: Train the model                                       # Training means the AI studies examples and learns from mistakes.
# Epoch means how many times the model studies the training data.# One epoch means one full pass through the training data.
# For camp, 5 epochs is enough.                                 # More epochs may improve learning, but they also take more time.
# ============================================================  # End of the Step 6 heading.

history = model.fit(                        # Train the model and save training results in a variable named history.
    x_train,                                # Give the model the training images.
    y_train,                                # Give the model the correct answers for the training images.
    epochs=5,                               # Train for 5 full passes through the training dataset.
    validation_split=0.1                    # Use 10% of the training data to check performance while training.
)                                           # End the training command.

print("Model training completed!")          # Print this after training is finished.
```

```python

# ============================================================  # Visual separator for the next section.
# Step 7: Test the model                                        # Testing checks how well the model works on images it did not train on.
# This checks how well the model performs on new images.         # New images help us see whether the AI really learned useful patterns.
# ============================================================  # End of the Step 7 heading.

test_loss, test_accuracy = model.evaluate(x_test, y_test)  # Test the model using test images and their correct answers.

print("Test Accuracy:", round(test_accuracy * 100, 2), "%")  # Convert accuracy to a percent and print it with 2 decimal places.
```

```python

# ============================================================  # Visual separator for the next section.
# Step 8: Plot accuracy and loss                                # Graphs help us understand how the model learned over time.
# Accuracy shows how often the model is correct.                 # Higher accuracy usually means better performance.
# Loss shows how much error the model makes.                     # Lower loss usually means the model is making fewer mistakes.
# ============================================================  # End of the Step 8 heading.

plt.figure(figsize=(8, 5))                  # Create a blank graph area that is 8 inches wide and 5 inches tall.
plt.plot(history.history["accuracy"], label="Training Accuracy")      # Draw the accuracy on the training data for each epoch.
plt.plot(history.history["val_accuracy"], label="Validation Accuracy") # Draw the accuracy on the validation data for each epoch.
plt.title("Model Accuracy")                 # Add a title to the accuracy graph.
plt.xlabel("Epoch")                         # Label the horizontal axis with the training epoch number.
plt.ylabel("Accuracy")                      # Label the vertical axis with the accuracy value.
plt.legend()                                # Show the labels so we know which line is which.
plt.show()                                  # Display the accuracy graph.

plt.figure(figsize=(8, 5))                  # Create a second blank graph area for loss.
plt.plot(history.history["loss"], label="Training Loss")          # Draw the training loss for each epoch.
plt.plot(history.history["val_loss"], label="Validation Loss")     # Draw the validation loss for each epoch.
plt.title("Model Loss")                     # Add a title to the loss graph.
plt.xlabel("Epoch")                         # Label the horizontal axis with the training epoch number.
plt.ylabel("Loss")                          # Label the vertical axis with the loss value.
plt.legend()                                # Show the labels for the two loss lines.
plt.show()                                  # Display the loss graph.
```

```python

# ============================================================  # Visual separator for the next section.
# Step 9: Make predictions on test images                       # Now the trained AI will guess digits from the test dataset.
# The model predicts which digit it sees.                        # A prediction is the model's best guess.
# ============================================================  # End of the Step 9 heading.

predictions = model.predict(x_test)         # Ask the model to predict a digit for every test image.

plt.figure(figsize=(12, 5))                 # Create a blank drawing area that is 12 inches wide and 5 inches tall.

for i in range(10):                         # Repeat the following indented lines for the first 10 test images.
    predicted_label = np.argmax(predictions[i])  # Choose the digit with the highest prediction score.
    actual_label = y_test[i]                # Get the correct answer for the same test image.

    plt.subplot(2, 5, i + 1)                # Make a 2 by 5 grid and choose the next box.
    plt.imshow(x_test[i], cmap="gray")      # Show the test image in black and white.
    plt.title(f"Predicted: {predicted_label}\nActual: {actual_label}")  # Show the AI guess and the correct answer.
    plt.axis("off")                         # Hide axis numbers so the image is cleaner.

plt.tight_layout()                          # Adjust spacing so the images and titles fit nicely.
plt.show()                                  # Display the prediction images.
```

```python

# ============================================================  # Visual separator for the next section.
# Step 10: Draw your own digit in Google Colab                  # This section lets students draw a digit using the mouse.
# This creates a simple drawing canvas.                         # A canvas is a drawing area inside the notebook.
# Students can draw a digit, and the AI will predict it.         # After drawing, the image is prepared for the trained model.
# ============================================================  # End of the Step 10 heading.

from IPython.display import HTML, display    # HTML lets us show a small webpage-like drawing tool inside the notebook.
from google.colab.output import eval_js      # eval_js can run JavaScript in Colab; it is imported here for Colab support.
from base64 import b64decode                 # b64decode helps convert the saved drawing from text form back into image bytes.
from PIL import Image                        # PIL helps Python open, resize, and edit images.
import io                                    # io helps Python treat image bytes like a file in memory.

canvas_html = """                            <!-- Start a long text block that contains HTML and JavaScript code. -->
<canvas id="canvas" width="280" height="280" style="border:2px solid black; background-color:white;"></canvas> <!-- Create a 280 by 280 drawing box. -->
<br>                                         <!-- Move the buttons to the next line. -->
<button onclick="clearCanvas()">Clear</button> <!-- Add a button that erases the drawing. -->
<button onclick="saveCanvas()">Predict Digit</button> <!-- Add a button that saves the drawing for prediction. -->

<script>                                     // Start JavaScript code that controls the drawing box.
var canvas = document.getElementById('canvas'); // Find the canvas element by its id name.
var ctx = canvas.getContext('2d');           // Create a 2D drawing tool for the canvas.
ctx.fillStyle = "white";                    // Choose white as the starting background color.
ctx.fillRect(0, 0, canvas.width, canvas.height); // Fill the whole canvas with white.

var drawing = false;                         // Keep track of whether the mouse is currently drawing.

canvas.addEventListener('mousedown', function(e) { // When the mouse button is pressed on the canvas, run this function.
    drawing = true;                          // Turn drawing mode on.
    ctx.beginPath();                         // Start a new drawing path so the line begins cleanly.
});                                          // End the mouse down function.

canvas.addEventListener('mouseup', function(e) { // When the mouse button is released, run this function.
    drawing = false;                         // Turn drawing mode off.
});                                          // End the mouse up function.

canvas.addEventListener('mousemove', function(e) { // When the mouse moves over the canvas, run this function.
    if (!drawing) return;                    // If drawing mode is off, stop here and do nothing.
    var rect = canvas.getBoundingClientRect(); // Get the canvas position on the screen.
    ctx.lineWidth = 18;                      // Make the drawing line thick so it looks like handwriting.
    ctx.lineCap = 'round';                   // Make the ends of the line rounded instead of square.
    ctx.strokeStyle = 'black';               // Draw using black ink.
    ctx.lineTo(e.clientX - rect.left, e.clientY - rect.top); // Draw a line to the current mouse position.
    ctx.stroke();                            // Actually paint the line on the canvas.
    ctx.beginPath();                         // Start a new path for smoother drawing.
    ctx.moveTo(e.clientX - rect.left, e.clientY - rect.top); // Move the pen to the current mouse position.
});                                          // End the mouse move function.

function clearCanvas() {                     // Define the function that clears the drawing.
    ctx.fillStyle = "white";                // Choose white as the erase color.
    ctx.fillRect(0, 0, canvas.width, canvas.height); // Paint the entire canvas white again.
}                                            // End the clearCanvas function.

function saveCanvas() {                      // Define the function that saves the drawing.
    var dataURL = canvas.toDataURL('image/png'); // Convert the canvas drawing into PNG image data.
    google.colab.kernel.invokeFunction('notebook.save_digit', [dataURL], {}); // Send the image data from JavaScript to Python.
}                                            // End the saveCanvas function.
</script>                                    <!-- End the JavaScript code. -->
"""                                          # End the long HTML and JavaScript text block.

drawn_image = None                           # Create a variable to store the student's drawn digit later.

def save_digit(data_url):                    # Define a Python function that receives the drawing from the canvas.
    global drawn_image                       # Tell Python we want to update the drawn_image variable created outside the function.

    binary = b64decode(data_url.split(",")[1]) # Remove the image header text and decode the actual image data.
    image = Image.open(io.BytesIO(binary)).convert("L") # Open the image and convert it to grayscale.

    # Resize to MNIST size: 28 x 28          # The model expects images to have the same size as MNIST images.
    image = image.resize((28, 28))           # Shrink the student's 280 by 280 drawing into a 28 by 28 image.

    # MNIST has white digits on black background. # The training images have bright digits and dark backgrounds.
    # Our canvas has black digit on white background, so we invert it. # We flip colors so the drawing matches MNIST better.
    image_array = np.array(image)            # Convert the image into a NumPy array of pixel numbers.
    image_array = 255 - image_array          # Invert the colors: black becomes white, and white becomes black.

    # Normalize                               # Scale pixel values so they match the 0 to 1 range used during training.
    image_array = image_array / 255.0        # Divide all pixel values by 255 to make them between 0 and 1.

    drawn_image = image_array                # Store the processed drawing so the next code cell can predict it.

    print("Digit saved! Now run the prediction section below.") # Tell the student the drawing is ready for prediction.

from google.colab import output              # Import Colab's output tool so JavaScript can call a Python function.
output.register_callback("notebook.save_digit", save_digit) # Connect the JavaScript save button to the Python save_digit function.

display(HTML(canvas_html))                   # Show the drawing canvas and buttons in the notebook.
```

```python

# ============================================================  # Visual separator for the next section.
# Step 11: Predict the digit you drew                           # Now the AI predicts the student's own handwritten digit.
# Run this after clicking "Predict Digit" on the canvas above.  # The drawing must be saved before this cell can use it.
# ============================================================  # End of the Step 11 heading.

if drawn_image is not None:                  # Check whether the student already drew and saved a digit.
    input_image = np.expand_dims(drawn_image, axis=0) # Add one extra dimension so the model sees this as one image in a batch.

    prediction = model.predict(input_image)  # Ask the trained model to predict the digit in the student's drawing.
    predicted_digit = np.argmax(prediction)  # Pick the digit with the highest prediction score.
    confidence = np.max(prediction)          # Get the highest score, which shows how confident the model is.

    plt.imshow(drawn_image, cmap="gray")     # Show the processed version of the student's drawing.
    plt.title(f"AI Prediction: {predicted_digit} | Confidence: {confidence:.2f}") # Show the predicted digit and confidence score.
    plt.axis("off")                          # Hide axis numbers because they are not useful here.
    plt.show()                               # Display the image and prediction title.

    print("The AI thinks you wrote:", predicted_digit) # Print the predicted digit in a simple sentence.
    print("Confidence:", round(confidence * 100, 2), "%") # Print the confidence as a percentage.
else:                                        # Run this part if no drawing has been saved yet.
    print("Please draw a digit above and click 'Predict Digit' first.") # Tell the student what to do before running this cell.
```

```python

# ============================================================  # Visual separator for the next section.
# Step 12: Save the trained model                               # Saving lets us keep the AI model after training.
# Students can keep their AI model.                             # The saved file can be downloaded or reused later.
# ============================================================  # End of the Step 12 heading.

model.save("handwritten_digit_recognition_model.h5") # Save the trained model into a file with the .h5 extension.

print("Model saved as handwritten_digit_recognition_model.h5") # Print the saved model filename.

# ============================================================  # Visual separator for the next section.
# Step 13: Reflection questions for presentation                # These questions help students explain what they built.
# Students can answer these in their final showcase.            # The showcase is where students present their project.
# ============================================================  # End of the Step 13 heading.

reflection_questions = [                    # Create a list that stores all reflection questions.
    "1. What problem does your AI model solve?", # Question 1 asks students to explain the project goal.
    "2. What dataset did you use?",         # Question 2 asks students to identify MNIST as the dataset.
    "3. How did the AI learn from handwritten digits?", # Question 3 asks students to describe training in their own words.
    "4. What was your model accuracy?",     # Question 4 asks students to report the test accuracy.
    "5. Did the model make any mistakes?",  # Question 5 asks students to think about incorrect predictions.
    "6. Why might some digits be harder to recognize than others?", # Question 6 asks students to reason about messy handwriting.
    "7. How could this type of AI be used in real life?" # Question 7 asks students to connect the project to real applications.
]                                           # End the list of reflection questions.

print("\nReflection Questions:")             # Print a blank line and then the heading for the questions.
for question in reflection_questions:        # Go through each question in the reflection_questions list.
    print(question)                          # Print the current question.
```
